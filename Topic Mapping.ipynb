```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.read_csv("datasets/tcga/keras/mainTable_test.csv", index_col=0)
```

```python
with open("corpus.txt",'w') as file:
    for sample in df.columns:
        for g in np.array(df[sample].sort_values(ascending=False).index[:100],dtype=str):
            file.write(g)
            file.write(" ")
        file.write("\n")
```

```python
df_words = pd.read_csv("topicmapping/corpus/word_wn_count.txt", sep=' ', header=None)
df_words.columns=['word', 'word-id', 'occurrence']
df_words.sort_values('word-id')
```

```python
df_topic_distr = pd.read_csv("topicmapping/corpus/lda_gammas_final.txt", sep=' ', header=None).drop(10,1)
df_topic_distr.index=df.columns
df_topic_distr.columns=['Topic %d'%(t+1) for t in df_topic_distr.columns]
df_topic_distr=df_topic_distr.transpose()
df_topic_distr
```

```python
clusters = df_topic_distr.values.argmax(0)
df_clusters=pd.DataFrame()
for cluster in range(np.max(clusters)):
    elems=df.columns[clusters==cluster].values
    df_clusters.insert(0,'Cluster %d'%cluster,np.concatenate([elems, ['' for _ in range(len(df.columns)- len(elems))]]))
```

```python
df_clusters.sort_index(axis=1).to_csv("tm_level_0_clusters.csv", index=False, header=True)
```
